```python
import matplotlib.pyplot as plt
import json
import statistics
import numpy as np
```

```python
def plot_accuracy(accuracies, y_lim, title, save=False, filename='accuracies'):
    y_labels = ['Test Accuracy', 'Training Accuracy']
    fig, axes = plt.subplots(1, 2, figsize=(13, 5))
    for i in range(2):
        axes[i].plot([1, 2, 3, 4, 5, 6, 7], accuracies[i][0], label='rf', marker='o', linestyle='-')
        axes[i].plot([1, 2, 3, 4, 5, 6, 7], accuracies[i][1], label='svmrad', marker='s', linestyle=(0, (1, 1)))
        axes[i].plot([1, 2, 3, 4, 5, 6, 7], accuracies[i][2], label='svmlin', marker='X', linestyle=(0, (3, 1, 1, 1, 1, 1)))
        axes[i].set_ylim(y_lim)
        axes[i].set_xticks([1, 2, 3, 4 , 5, 6, 7])
        axes[i].set_xticklabels(['k=1', 'k=2', 'k=3', 'k=4', 'k=5', 'k=6', 'k=7'])
        axes[i].set_ylabel(y_labels[i])
        axes[i].legend(loc='lower right')
    plt.subplots_adjust(wspace=0.4)
    plt.suptitle(title)
    if save:
        plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()
```

```python
def plot_errors(x, ylim, title, save=False, filename='errors'):
    y_labels = ['Test - Estimation Distribution', 'Training - Estimation Distribution']
    fig, axes = plt.subplots(1, 2, figsize=(13, 5))
    for i in range(2):
        axes[i].boxplot(x[i])
        axes[i].set_ylim(ylim)
        axes[i].set_xticks([1, 2, 3, 4 , 5, 6, 7])
        axes[i].set_xticklabels(['k=1', 'k=2', 'k=3', 'k=4', 'k=5', 'k=6', 'k=7'])
        axes[i].set_ylabel(y_labels[i])
    plt.subplots_adjust(wspace=0.4)
    plt.suptitle(title)
    if save:
        plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()
```

```python
def get_accuracy(data):
    mean_accs = {}
    for k in sorted(data):
        for model in data[k]: 
            mean_accs.setdefault(model, [])
            mean_accs[model].append(statistics.mean(data[k][model]) * 100)
    return mean_accs
```

```python
def get_errors(data):
    error_scores = {}
    for k in data:
        for model in data[k]: 
            error_scores.setdefault(model, {})
            error_scores[model].setdefault(k, 0)
            error_scores[model][k] = data[k][model]
    return error_scores
```

```python
def get_variance(dct):
    dist_matrix = []
    for k in sorted(dct):
        dist_matrix.append([i*100 for i in dct[k]])
    return dist_matrix
```

```python
with open('hierarchical class acc.json', 'r') as fr:
    data = json.load(fr)
    class_acc = get_accuracy(data)
```

```python
acc_list = [[class_acc['rf test'], class_acc['svmrad test'], class_acc['svmlin test']],
            [class_acc['rf train'], class_acc['svmrad train'], class_acc['svmlin train']]]
plot_accuracy(acc_list, (65, 102), 'Taxonomic Class Predictor', True, 'taxonomic class - merged dataset')
```

```python
with open('hierarchical class acc.json', 'r') as fr:
    data = json.load(fr)
    class_error = get_errors(data)
```

```python
rf_test = get_variance(class_error['rf test'])
rf_train = get_variance(class_error['rf train'])
plot_errors([rf_test, rf_train], (95, 100.5), 'Taxonomic Class Estimation Distribution by Random Forest', True, 'rf_est_dist')
```

```python
rf_test = get_variance(class_error['svmrad test'])
rf_train = get_variance(class_error['svmrad train'])
plot_errors([rf_test, rf_train], (60, 100.5), 'Taxonomic Class Estimation Distribution by SVM - Radial Kernel', True, 'svmrad_est_dist')
```

```python
rf_test = get_variance(class_error['svmlin test'])
rf_train = get_variance(class_error['svmlin train'])
plot_errors([rf_test, rf_train], (85, 100.5), 'Taxonomic Class Estimation Distribution by SVM - Linear Kernel', True, 'svmlin_est_dist')
```

```python
with open('non-hierarchical acc score.json', 'r') as fr:
    data = json.load(fr)
    chirop_results = get_accuracy(data['Chiroptera'])
    rodent_results = get_accuracy(data['Rodentia'])
    aves_results = get_accuracy(data['Aves'])
```
