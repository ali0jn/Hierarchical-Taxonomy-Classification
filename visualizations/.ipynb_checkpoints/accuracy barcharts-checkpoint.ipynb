```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 5
svmlin_test = [90.39, 97.15, 97.94, 96.98, 87.81]
svmrad_test = [84.85, 95.70, 97.00, 94.00, 79.57]
rf_test = [88.98, 96.73, 97.52, 95.87, 83.10]

fig, ax = plt.subplots(figsize=(12, 8.6))

ind = np.arange(N)    # the x locations for the groups
width = 0.25         # the width of the bars
ax.bar(ind, svmlin_test, width, bottom=0, label='svmlin')
for i, v in enumerate(svmlin_test):
    ax.text(i-0.12, v+1, str(v))
    
ax.bar(ind + width, svmrad_test, width, bottom=0, label='svmrad')
for i, v in enumerate(svmrad_test):
    ax.text(i+width-0.1, v+1, str(v))

ax.bar(ind + width*2, rf_test, width, bottom=0, label='rf')
for i, v in enumerate(rf_test):
    ax.text(i+2*width-0.12, v+1, str(v))

ax.set_title('Maximum Accuracy of Models', {'fontsize': 15})
ax.set_xticks(ind + width*2 / 2)
ax.set_xticklabels(('Polypodiopsida', 'Pucciniomycetes', 'Chiroptera', 'Rodentia', 'Aves'), {'fontsize': 13})
ax.set_ylabel('Accuracy %', {'fontsize': 13})
ax.set_yticks(np.arange(0, 101, 10))
ax.legend(prop={'size': 12})
plt.savefig('non_hierarchical_score_chart', bbox_inches='tight', dpi=300)
plt.show()
```

```python
x = np.arange(3)
test_acc = [90.7, 93.00, 91.00]

fig, ax = plt.subplots(figsize=(10, 8))
ax.bar(x, test_acc, label='Testing Accuracy')
for i, v in enumerate(test_acc):
    plt.text(i-0.08, v+1, str(v))
    
plt.xticks(x, ('non-hierarchical RF', 'hierarchical Linear SVM', 'hierarchical RF'), fontsize= 13)
plt.legend(prop={'size': 11})
plt.yticks(np.arange(0, 110, 10))
plt.title('Comparison: Hierarchical vs Non-hierarchical', {'fontsize': 15})
plt.savefig('hierarchical_score_chart', bbox_inches='tight', dpi=300)
plt.show()
```
